```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
from eu_survey_correlation.surveys.ess_scraper import ESSCodebookParser
import os
```

```python
ess_datapath = "data/surveys/ess"
all_questions_df = pd.DataFrame()
for ess_dir in os.listdir(ess_datapath):
    dir_path = ess_datapath + f"/{ess_dir}"
    for filename in os.listdir(dir_path):
        if filename.endswith(".html"):
            filepath = dir_path + f"/{filename}"
            parser = ESSCodebookParser(filepath)
            df = parser.parse()
            df["ess_id"] = ess_dir
            all_questions_df = pd.concat([all_questions_df, df]).reset_index(drop=True)
```
