```python
import pandas as pd
from pathlib import Path
from tqdm import tqdm
```

## Get all questions and sheet_id

```python
# Load an Excel file and explore its structure
content_dfs = []
excel_path = Path("data/volume_b_docs")
for file in tqdm(excel_path.glob("*.xlsx")):
    xl_file = pd.ExcelFile(file)
    first_sheet_name = xl_file.sheet_names[0]
    # skip the rows until the 2nd column is not empty
    content_df = pd.read_excel(xl_file, sheet_name=first_sheet_name)
    content_df = content_df[content_df.iloc[:, 1].notna()]
    content_df = content_df[content_df.iloc[:, 1].notna()]
    content_df.columns = content_df.iloc[0]
    content_df = content_df.iloc[1:]

    # check if first col first row == "Wave:". if so, skip 2 rows
    if content_df.iloc[0, 0] == "Fieldwork:":
        content_df = content_df.iloc[2:]
        content_df.columns = ["sheet_id", "question_fr", "question_en"]
    else:
        content_df.columns = ["sheet_id", "question_en"]
    content_df["file_name"] = file.name
    content_dfs.append(content_df)
```

```python
all_survey_questions_df = pd.concat(content_dfs)
```

```python
all_survey_questions_df = all_survey_questions_df[
    ["sheet_id", "question_en", "question_fr", "file_name"]
].copy()

all_survey_questions_df.to_csv("./data/surveys/all_survey_questions.csv", index=False)
```
