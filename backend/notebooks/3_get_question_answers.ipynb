```python
import pandas as pd
import os
from pathlib import Path
```

```python
def find_header_row(xl: pd.ExcelFile, sheet_name: str) -> int:
    """
    Find the header row where the 2nd column contains 'EU27'.
    Returns the index of the first row with 'EU27' in the 2nd column.
    """
    # Read without headers to scan for EU27
    raw_df = pd.read_excel(xl, sheet_name=sheet_name, header=None, nrows=200)

    # Find rows where 2nd column (index 1) equals "EU27"
    eu27_rows = raw_df[raw_df.iloc[:, 1] == "EU27"].index.tolist()

    if not eu27_rows:
        # maybe it is the type of files with "UE27\n-\nUE27"
        if "UE27\n-\nUE27" in raw_df.iloc[:, 2].values:
            print("OUI")
            return [7, 8]
    # Return the first occurrence
    return eu27_rows


def read_eurobarometer_sheet(xl: pd.ExcelFile, sheet_name: str) -> pd.DataFrame:
    """
    Read a Eurobarometer Excel sheet with multi-row headers.
    Dynamically finds header row where 2nd column has 'EU27'.
    """
    # Find header row dynamically
    header_rows = find_header_row(xl, sheet_name)

    # Read with the detected header row
    df = pd.read_excel(
        xl,
        sheet_name=sheet_name,
        header=header_rows,
    )

    # Flatten multi-level column names
    if df.columns.nlevels == 2:
        new_columns = []
        for col in df.columns:
            # col is a tuple like ('Sexe\n-\nGender', 'Homme\n-\nMan')
            level0 = (
                str(col[0]) if pd.notna(col[0]) and "Unnamed" not in str(col[0]) else ""
            )
            level1 = (
                str(col[1]) if pd.notna(col[1]) and "Unnamed" not in str(col[1]) else ""
            )

            # Clean up newlines
            level0 = level0.replace("\n", " ").strip()
            level1 = level1.replace("\n", " ").strip()

            # Combine: "Category | Subcategory"
            if level0 and level1:
                col_name = f"{level0} | {level1}"
            elif level1:
                col_name = level1
            elif level0:
                col_name = level0
            else:
                col_name = f"col_{len(new_columns)}"

            new_columns.append(col_name)

        df.columns = new_columns

    # Remove rows that are all NaN
    df = df.dropna(how="all")

    return df
```

```python
all_survey_questions_df = pd.read_csv("./data/surveys/all_survey_questions.csv")
```

```python
# for file in all_survey_questions_df["file_name"].unique():
for file in ["ebs_504_volume_B.xlsx"]:
    # for file in ["eb_90_volume_B.xls"]:
    related_df = all_survey_questions_df[all_survey_questions_df["file_name"] == file]
    xl_file = pd.ExcelFile(Path("./data/volume_b_docs") / file)

    for sheet_id in related_df["sheet_id"]:
        sheet_df = read_eurobarometer_sheet(xl_file, sheet_id)
```
