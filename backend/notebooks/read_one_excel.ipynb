```python
def read_eurobarometer_sheet(xl: pd.ExcelFile, sheet_name: str) -> pd.DataFrame:
    """
    Read a Eurobarometer Excel sheet with multi-row headers.
    Combines header rows 6 and 7 into proper column names.
    """
    # Read with multi-row header (rows 6 and 7)
    df = pd.read_excel(xl, sheet_name=sheet_name, header=[7, 8])

    # Flatten multi-level column names
    new_columns = []
    for col in df.columns:
        # col is a tuple like ('Sexe\n-\nGender', 'Homme\n-\nMan')
        level0 = (
            str(col[0]) if pd.notna(col[0]) and "Unnamed" not in str(col[0]) else ""
        )
        level1 = (
            str(col[1]) if pd.notna(col[1]) and "Unnamed" not in str(col[1]) else ""
        )

        # Clean up newlines
        level0 = level0.replace("\n", " ").strip()
        level1 = level1.replace("\n", " ").strip()

        # Combine: "Category | Subcategory"
        if level0 and level1:
            col_name = f"{level0} | {level1}"
        elif level1:
            col_name = level1
        elif level0:
            col_name = level0
        else:
            col_name = f"col_{len(new_columns)}"

        new_columns.append(col_name)

    df.columns = new_columns

    # Remove rows that are all NaN
    df = df.dropna(how="all")

    return df
```

```python
# Load an Excel file and explore its structure
excel_path = Path("./excels/Lien_vers_Sport_et_Physical_Activity_EBS525_volume_B.xlsx")

# Get all sheet names
xl = pd.ExcelFile(excel_path)
print(f"Sheet names: {xl.sheet_names}")
```
