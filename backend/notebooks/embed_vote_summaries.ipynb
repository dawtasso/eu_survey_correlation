```python
from google import genai
from dotenv import load_dotenv
import pandas as pd
from tqdm import tqdm
import os

load_dotenv()
```

```python
df_summaries = pd.read_csv("data/votes/vote_summaries.csv")
df_votes = pd.read_csv("data/votes/votes.csv")
```

```python
unique_summaries = df_summaries["summary"].unique().tolist()
```

```python
genai_client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))
BATCH_SIZE = 100
```

```python
embedding_dict = {}
for i in tqdm(range(0, len(unique_summaries), BATCH_SIZE)):
    batch = unique_summaries[i : i + BATCH_SIZE]
    result = genai_client.models.embed_content(
        model="gemini-embedding-001",
        contents=batch,
    )
    for summary, embedding_obj in zip(batch, result.embeddings):
        embedding_dict[summary] = embedding_obj.values
```
